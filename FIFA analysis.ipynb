```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
import csv 
file = []
with open("FIFA-21-Complete.csv","r",encoding = "utf-8") as f:
  data_reader = csv.reader(f,delimiter=";")
  #header = next(data_reader)
  for i in data_reader:
    file = [row for row in data_reader]
```

```python
name = []
country = []
position = []
age = []
potential = []
team = []
overall = []
for i in range(0,len(file)):
    name.append(file[i][1])
    country.append(file[i][2])
    position.append(file[i][3])
    age.append(int(file[i][5]))
    potential.append(int(file[i][7]))
    team.append(file[i][8])
    overall.append(int(file[i][4]))
```

```python
name = np.array(name)
country = np.array(country)
position = np.array(position)
age = np.array(age)
potential = np.array(potential)
team = np.array(team)
overall = np.array(overall)
```

```python
# Vẽ phân phối potential theo team
fig,ax = plt.subplots(figsize=(12,6), num=1)
ax.hist(potential, bins=np.arange(potential.min(),potential.max()), alpha=0.75, width=0.8,
                          color='red', edgecolor='black', linewidth=1.0)
plt.axvline(potential.mean(), color = "g", ls = "--", label = "Mean Of Potential")
xtick_labels = np.arange(potential.min(),potential.max())
ax.set_xticks(xtick_labels+0.4)
ax.set_xticklabels(xtick_labels)
ax.set_xlabel("Potential")
plt.show()
```

```python
# Top 10 cầu thủ có potential cao nhất
vitri = []
top10_potential = np.sort(potential)[-10:]
a = np.argsort(potential)[-10:]

top10_name = name[a]
for i in top10_name:
    vt = np.argwhere(name == i)
    vitri.append(vt)
    vi_tri = np.squeeze(vitri)
top10_overall = overall[vi_tri]
top10_age = age[vi_tri]

plt.figure(figsize=(12,8))
plt.barh(top10_name,top10_potential)
plt.barh(top10_name,top10_overall)
plt.barh(top10_name,top10_age)
plt.legend(["Potential","Overall","Age"])
```

```python
# ClB nào có trung bình potential cao nhất
```

```python
arg_potential = []
clubs = np.unique(team)
for i in clubs:
    a = np.argwhere(i == team)
    b = potential[a]
    c = round(b.mean(),2)
    arg_potential.append(c)
top10_potential_team = np.sort(arg_potential)[-10:]
vitri = np.argsort(arg_potential)[-10:]

# Vẽ biểu đồ
plt.barh(team[vitri],top10_potential_team)
plt.axvline(potential.mean(), color = "r", ls = "--", label = "Mean Of Potential")
```

```python
# Tạo 1 team có các vị trí theo potential cao nhất
```

```python
# Tách các vị trí ra
positions = []
for i in position:
    split_position = i.split("|")
    position = split_position[0]
    positions.append(position)
positions = np.array(positions)
```

```python
def vitri(x,y):
    v = []
    vitri_potential = []
    b = np.where(x == positions) 
    b = np.squeeze(b)
    m = np.where(y == team)
    m = np.squeeze(m)
    
    for e in b:
        for h in m:
            if e == h:
                v.append(e)
    
    for i in v:
        potential_position = potential[i]
        vitri_potential.append(potential_position)
    vitri_potential = np.array(vitri_potential)
# Tìm vị trí có potential cao nhất và có overall cao nhất
    top1_potential = np.sort(vitri_potential)[-1:]

    c = np.where(top1_potential == potential) and v
    
    if x == "CB":
        d = c[:2]
    elif x == "CM":
        d = c[:2]
    else:
        d = c[:1]
    top1_potential_name = name[d]
    print(top1_potential_name)
```

```python
vitri("LB","Manchester United ")
```
